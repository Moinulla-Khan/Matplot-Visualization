```python
plt.figure(figsize=(20, 8))
sns.barplot(x='category', y='profit', data=df)
plt.title('Profit based on category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(20, 8))
sns.barplot(x='category', y='sales', data=df)
plt.title('Sales based on category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```
